#### 15. Write a python function sin (x, n) to calculate the value of sin(x) using its Taylor series expansion up to n terms. 

```python
def sin(x, n):
    result = 0
    term = x

    for i in range(n):
        result += term
        term *= -x**2 / ((2*i + 2) * (2*i + 3))

    return result

# Ask input from user
x = float(input("Enter the value of x (in radians): "))
n = int(input("Enter the number of terms n: "))

print(f"sin({x}) ≈ {sin(x, n)}")
```

#### 16. Write a program to determine EOQ using various inventory models.

```python
import math

# Base Class
class InventoryModel:
    def __init__(self):
        self.D = float(input("Enter Demand (D): "))
        self.S = float(input("Enter Ordering Cost (S): "))
        self.H = float(input("Enter Holding Cost (H): "))

    def calculate_eoq(self):
        pass


# Derived Class 1: Basic EOQ
class BasicEOQ(InventoryModel):
    def calculate_eoq(self):
        eoq = math.sqrt((2 * self.D * self.S) / self.H)
        print(f"\nBasic EOQ = {eoq:.2f}")


# Derived Class 2: EOQ with Shortages
class EOQWithShortage(InventoryModel):
    def __init__(self):
        super().__init__()
        self.P = float(input("Enter Shortage Cost (P): "))

    def calculate_eoq(self):
        eoq = math.sqrt((2 * self.D * self.S / self.H) * ((self.H + self.P) / self.P))
        print(f"\nEOQ with Shortage = {eoq:.2f}")


# Derived Class 3: Production EOQ (EPQ Model)
class ProductionEOQ(InventoryModel):
    def __init__(self):
        super().__init__()
        self.p = float(input("Enter Production Rate (p): "))
        self.d = float(input("Enter Demand Rate (d): "))

    def calculate_eoq(self):
        if self.p <= self.d:
            print("\nProduction rate must be greater than demand rate!")
            return
        
        eoq = math.sqrt((2 * self.D * self.S / self.H) * (self.p / (self.p - self.d)))
        print(f"\nProduction EOQ (EPQ) = {eoq:.2f}")


# Main Program
def main():
    while True:
        print("\n===== EOQ MODEL MENU =====")
        print("1. Basic EOQ")
        print("2. EOQ with Shortage")
        print("3. Production EOQ (EPQ)")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            model = BasicEOQ()
            model.calculate_eoq()

        elif choice == '2':
            model = EOQWithShortage()
            model.calculate_eoq()

        elif choice == '3':
            model = ProductionEOQ()
            model.calculate_eoq()

        elif choice == '4':
            print("Exiting program...")
            break

        else:
            print("Invalid choice! Please try again.")

main()
```

#### 17. Write a program to determine different characteristics using various queuing models.

```python
import math

# Base Class
class QueueModel:
    def __init__(self):
        self.lam = float(input("Enter Arrival Rate (λ): "))
        self.mu = float(input("Enter Service Rate (μ): "))

    def calculate(self):
        pass


# M/M/1 Model
class MM1(QueueModel):
    def calculate(self):
        if self.lam >= self.mu:
            print("\nSystem is unstable (λ ≥ μ)")
            return
        
        L = self.lam / (self.mu - self.lam)
        Lq = (self.lam ** 2) / (self.mu * (self.mu - self.lam))
        W = 1 / (self.mu - self.lam)
        Wq = self.lam / (self.mu * (self.mu - self.lam))

        print("\n--- M/M/1 Results ---")
        print(f"L  = {L:.3f}")
        print(f"Lq = {Lq:.3f}")
        print(f"W  = {W:.3f}")
        print(f"Wq = {Wq:.3f}")


# M/M/1/K Model (Finite Capacity)
class MM1K(QueueModel):
    def __init__(self):
        super().__init__()
        self.K = int(input("Enter System Capacity (K): "))

    def calculate(self):
        rho = self.lam / self.mu

        if rho == 1:
            P0 = 1 / (self.K + 1)
        else:
            P0 = (1 - rho) / (1 - rho ** (self.K + 1))

        L = (rho * (1 - (self.K + 1) * rho**self.K + self.K * rho**(self.K + 1))) / ((1 - rho) * (1 - rho**(self.K + 1)))
        Lq = L - (1 - P0)

        W = L / (self.lam * (1 - (rho**self.K * P0)))
        Wq = Lq / (self.lam * (1 - (rho**self.K * P0)))

        print("\n--- M/M/1/K Results ---")
        print(f"L  = {L:.3f}")
        print(f"Lq = {Lq:.3f}")
        print(f"W  = {W:.3f}")
        print(f"Wq = {Wq:.3f}")


# M/M/c Model (Multiple Servers)
class MMC(QueueModel):
    def __init__(self):
        super().__init__()
        self.c = int(input("Enter Number of Servers (c): "))

    def calculate(self):
        rho = self.lam / (self.c * self.mu)

        if rho >= 1:
            print("\nSystem is unstable (ρ ≥ 1)")
            return

        # Calculate P0
        sum_terms = sum([(self.lam / self.mu) ** n / math.factorial(n) for n in range(self.c)])
        last_term = ((self.lam / self.mu) ** self.c) / (math.factorial(self.c) * (1 - rho))
        P0 = 1 / (sum_terms + last_term)

        # Lq
        Lq = (P0 * ((self.lam / self.mu) ** self.c) * rho) / (math.factorial(self.c) * (1 - rho) ** 2)

        L = Lq + (self.lam / self.mu)
        Wq = Lq / self.lam
        W = Wq + (1 / self.mu)

        print("\n--- M/M/c Results ---")
        print(f"L  = {L:.3f}")
        print(f"Lq = {Lq:.3f}")
        print(f"W  = {W:.3f}")
        print(f"Wq = {Wq:.3f}")


# Main
def main():
    while True:
        print("\n===== QUEUING MODEL MENU =====")
        print("1. M/M/1 Model")
        print("2. M/M/1/K Model")
        print("3. M/M/c Model")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            model = MM1()
            model.calculate()

        elif choice == '2':
            model = MM1K()
            model.calculate()

        elif choice == '3':
            model = MMC()
            model.calculate()

        elif choice == '4':
            print("Exiting program...")
            break

        else:
            print("Invalid choice! Try again.")


main()
```

#### 18. Write a program to implement Inheritance. Create a class Employee inherit two classes Manager and Clerk from Employee.

#### 19. Write a program to fit Poisson distribution on a given data. 

```python
import math

def factorial(n):
    return math.factorial(n)

# Main
def main():
    print("Fit Poisson Distribution")

    # Input number of observations
    n = int(input("Enter number of observations: "))

    # Input data values
    data = []
    print("Enter observed values:")
    for i in range(n):
        value = int(input(f"Value {i+1}: "))
        data.append(value)

    # Calculate mean (lambda)
    lam = sum(data) / n
    print(f"\nCalculated Mean (λ) = {lam:.4f}")

    # Count frequency of each value
    freq = {}
    for value in data:
        freq[value] = freq.get(value, 0) + 1

    N = n  # total observations

    print("\nX\tObserved\tExpected")

    # Calculate expected frequencies
    for x in sorted(freq.keys()):
        px = (math.exp(-lam) * (lam ** x)) / factorial(x)
        expected = N * px

        print(f"{x}\t{freq[x]}\t\t{expected:.4f}")

main()
```

#### 20. Write a program to implement linear regression using python.

```python
def get_inputs():
    print("-" * 40)
    print("       LINEAR REGRESSION PROGRAM")
    print("-" * 40)

    while True:
        try:
            n = int(input("\nEnter number of observations: "))
            if n < 2:
                print("  Need at least 2 data points.")
                continue
            break
        except ValueError:
            print("  Please enter a valid integer.")

    x, y = [], []

    print("\nEnter X values:")
    for i in range(n):
        while True:
            try:
                x.append(float(input(f"  X[{i+1}]: ")))
                break
            except ValueError:
                print("  Invalid input. Enter a number.")

    print("\nEnter Y values:")
    for i in range(n):
        while True:
            try:
                y.append(float(input(f"  Y[{i+1}]: ")))
                break
            except ValueError:
                print("  Invalid input. Enter a number.")

    return n, x, y


def compute_regression(n, x, y):
    sum_x  = sum(x)
    sum_y  = sum(y)
    sum_xy = sum(x[i] * y[i] for i in range(n))
    sum_x2 = sum(xi ** 2 for xi in x)

    denominator = n * sum_x2 - sum_x ** 2
    if denominator == 0:
        raise ValueError("All X values are identical — regression is undefined.")

    slope     = (n * sum_xy - sum_x * sum_y) / denominator
    intercept = (sum_y / n) - slope * (sum_x / n)

    return slope, intercept


def compute_r_squared(n, x, y, slope, intercept):
    y_mean  = sum(y) / n
    ss_tot  = sum((yi - y_mean) ** 2 for yi in y)
    ss_res  = sum((y[i] - (intercept + slope * x[i])) ** 2 for i in range(n))

    return 1 - (ss_res / ss_tot) if ss_tot != 0 else 1.0


def display_results(slope, intercept, r_squared):
    print("\n" + "-" * 40)
    print("         REGRESSION RESULTS")
    print("-" * 40)
    print(f"  Intercept (a) : {intercept:.4f}")
    print(f"  Slope     (b) : {slope:.4f}")
    print(f"  Equation      : y = {intercept:.4f} + {slope:.4f}x")
    print(f"  R² Score      : {r_squared:.4f}  ({r_squared*100:.2f}% variance explained)")
    print("-" * 40)


def predict(slope, intercept):
    print("\n--- Prediction ---")
    while True:
        try:
            x_new = float(input("Enter X value to predict Y: "))
            break
        except ValueError:
            print("  Invalid input. Enter a number.")

    y_pred = intercept + slope * x_new
    print(f"  Predicted Y = {y_pred:.4f}")
    return y_pred


def main():
    n, x, y = get_inputs()

    try:
        slope, intercept = compute_regression(n, x, y)
    except ValueError as e:
        print(f"\nError: {e}")
        return

    r_squared = compute_r_squared(n, x, y, slope, intercept)
    display_results(slope, intercept, r_squared)

    while True:
        predict(slope, intercept)
        again = input("\nPredict another value? (y/n): ").strip().lower()
        if again != 'y':
            break

    print("\nThank you for using the Linear Regression Program!")


if __name__ == "__main__":
    main()
```

#### 21. Write a program to perform read and write operation with .csv file.

```python
import csv

def write_csv():
    filename = input("Enter file name (with .csv): ")

    n = int(input("Enter number of records: "))

    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)

        headers = input("Enter column names (comma-separated): ").split(',')
        writer.writerow(headers)

        for i in range(n):
            row = input(f"Enter data for row {i+1} (comma-separated): ").split(',')
            writer.writerow(row)

    print("\nData written successfully!")


def read_csv():
    filename = input("Enter file name to read (with .csv): ")

    try:
        with open(filename, mode='r') as file:
            reader = csv.reader(file)

            print("\n--- File Contents ---")
            for row in reader:
                print(row)

    except FileNotFoundError:
        print("File not found!")


def main():
    while True:
        print("\n===== CSV FILE MENU =====")
        print("1. Write to CSV")
        print("2. Read from CSV")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            write_csv()

        elif choice == '2':
            read_csv()

        elif choice == '3':
            print("Exiting program...")
            break

        else:
            print("Invalid choice! Try again.")


main()
```

```python
def list_program():
    print("--- LIST OPERATIONS ---")

    n = int(input("Enter number of elements: "))
    lst = []

    for i in range(n):
        val = input(f"Enter element {i+1}: ")
        lst.append(val)

    print("\nInitial List:", lst)

    while True:
        print("\n1. Append")
        print("2. Insert")
        print("3. Remove")
        print("4. Update")
        print("5. Display")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            val = input("Enter value to append: ")
            lst.append(val)

        elif choice == '2':
            pos = int(input("Enter position: "))
            val = input("Enter value: ")
            lst.insert(pos, val)

        elif choice == '3':
            val = input("Enter value to remove: ")
            if val in lst:
                lst.remove(val)
            else:
                print("Value not found!")

        elif choice == '4':
            pos = int(input("Enter index to update: "))
            val = input("Enter new value: ")
            lst[pos] = val

        elif choice == '5':
            print("List:", lst)

        elif choice == '6':
            break

        else:
            print("Invalid choice!")

list_program()
```

```python
def tuple_program():
    print("--- TUPLE OPERATIONS ---")

    n = int(input("Enter number of elements: "))
    temp = []

    for i in range(n):
        val = input(f"Enter element {i+1}: ")
        temp.append(val)

    tup = tuple(temp)
    print("\nInitial Tuple:", tup)

    while True:
        print("\n1. Add Element")
        print("2. Remove Element")
        print("3. Update Element")
        print("4. Display")
        print("5. Exit")

        choice = input("Enter choice: ")

        temp = list(tup)

        if choice == '1':
            val = input("Enter value to add: ")
            temp.append(val)

        elif choice == '2':
            val = input("Enter value to remove: ")
            if val in temp:
                temp.remove(val)
            else:
                print("Value not found!")

        elif choice == '3':
            pos = int(input("Enter index: "))
            val = input("Enter new value: ")
            temp[pos] = val

        elif choice == '4':
            print("Tuple:", tup)
            continue

        elif choice == '5':
            break

        else:
            print("Invalid choice!")
            continue

        tup = tuple(temp)
        print("Updated Tuple:", tup)

tuple_program()
```

```python
def dict_program():
    print("---- DICTIONARY OPERATIONS ----")

    n = int(input("Enter number of key-value pairs: "))
    d = {}

    for i in range(n):
        key = input(f"Enter key {i+1}: ")
        value = input(f"Enter value for {key}: ")
        d[key] = value

    print("\nInitial Dictionary:", d)

    while True:
        print("\n1. Add/Update")
        print("2. Delete")
        print("3. Display")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            key = input("Enter key: ")
            value = input("Enter value: ")
            d[key] = value

        elif choice == '2':
            key = input("Enter key to delete: ")
            if key in d:
                del d[key]
            else:
                print("Key not found!")

        elif choice == '3':
            print("Dictionary:", d)

        elif choice == '4':
            break

        else:
            print("Invalid choice!")

dict_program()
```

#### 27. Write programs to find correlation between dependent and independent variables. 

```python
import pandas as pd

def main():
    print("---- Correlation using Pandas ----")

    n = int(input("Enter number of observations: "))

    x = []
    y = []

    print("\nEnter X values:")
    for i in range(n):
        x.append(float(input(f"X[{i+1}]: ")))

    print("\nEnter Y values:")
    for i in range(n):
        y.append(float(input(f"Y[{i+1}]: ")))

    df = pd.DataFrame({
        "X": x,
        "Y": y
    })

    r = df["X"].corr(df["Y"])

    print("\nCorrelation Coefficient (r) =", round(r, 4))

main()
```

#### 28. Write programs for data visualization (Charts using plot () function, Pie Chart, Scatter Plot, Histogram, Bar Chart). 

```python
import matplotlib.pyplot as plt

def line_chart():
    n = int(input("Enter number of points: "))
    x, y = [], []

    print("Enter X values:")
    for i in range(n):
        x.append(float(input(f"X[{i+1}]: ")))

    print("Enter Y values:")
    for i in range(n):
        y.append(float(input(f"Y[{i+1}]: ")))

    plt.plot(x, y)
    plt.title("Line Chart")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid()
    plt.show()


def pie_chart():
    n = int(input("Enter number of categories: "))
    labels = []
    sizes = []

    for i in range(n):
        labels.append(input(f"Label {i+1}: "))
        sizes.append(float(input(f"Value for {labels[i]}: ")))

    plt.pie(sizes, labels=labels, autopct='%1.1f%%')
    plt.title("Pie Chart")
    plt.show()


def scatter_plot():
    n = int(input("Enter number of points: "))
    x, y = [], []

    for i in range(n):
        x.append(float(input(f"X[{i+1}]: ")))
        y.append(float(input(f"Y[{i+1}]: ")))

    plt.scatter(x, y)
    plt.title("Scatter Plot")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()


def histogram():
    n = int(input("Enter number of data points: "))
    data = []

    for i in range(n):
        data.append(float(input(f"Value {i+1}: ")))

    plt.hist(data)
    plt.title("Histogram")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()


def bar_chart():
    n = int(input("Enter number of categories: "))
    categories = []
    values = []

    for i in range(n):
        categories.append(input(f"Category {i+1}: "))
        values.append(float(input(f"Value for {categories[i]}: ")))

    plt.bar(categories, values)
    plt.title("Bar Chart")
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.show()


def main():
    while True:
        print("\n------- DATA VISUALIZATION MENU -------")
        print("1. Line Chart (plot)")
        print("2. Pie Chart")
        print("3. Scatter Plot")
        print("4. Histogram")
        print("5. Bar Chart")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            line_chart()
        elif choice == '2':
            pie_chart()
        elif choice == '3':
            scatter_plot()
        elif choice == '4':
            histogram()
        elif choice == '5':
            bar_chart()
        elif choice == '6':
            print("Exiting program...")
            break
        else:
            print("Invalid choice! Try again.")

main()
```

```python
from pulp import *

model = LpProblem("Maximize_Profit", LpMaximize)

x = LpVariable('x', lowBound=0)
y = LpVariable('y', lowBound=0)

model += 3 * x + 5 * y

model += 2 * x + 3 * y <= 12
model += x + y <= 5

model.solve()

print("Status:", LpStatus[model.status])
print("Optimal x =", x.varValue)
print("Optimal y =", y.varValue)
print("Maximum Value =", value(model.objective))
```

##### Program 1: Optimization (Minimize Function)

```python
from scipy.optimize import minimize

def func(x):
    return x**2 + 4*x + 4

x0 = 0

result = minimize(func, x0)

print("Optimal x =", result.x[0])
print("Minimum value =", result.fun)
```

##### Program 2: Integration (Area Under Curve)

```python
from scipy.integrate import quad

def f(x):
    return x**2
    
result, error = quad(f, 0, 2)

print("Integral value =", result)
print("Error estimate =", error)
```

```python
# Import PuLP library
from pulp import LpMaximize, LpProblem, LpVariable, value

# Create the model
model = LpProblem(name="LP_Model", sense=LpMaximize)

# Decision variables
A = LpVariable(name="A", lowBound=0)
B = LpVariable(name="B", lowBound=0)

# Objective function
model += 50 * A + 40 * B, "Maximize_Profit"

# Constraints
model += 2 * A + B <= 100, "Constraint_1"
model += A + 2 * B <= 80, "Constraint_2"

# Solve the model
model.solve()

# Print results
print("Status:", model.status)
print("Optimal value of A =", A.value())
print("Optimal value of B =", B.value())
print("Maximum value of Z =", value(model.objective))
```
